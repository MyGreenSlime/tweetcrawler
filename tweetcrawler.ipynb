```python
import tweepy
import json
```

```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

# Creating the authentication object
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
# Setting your access token and secret
auth.set_access_token(access_token, access_secret)

# Creating the API object while passing in auth information
api = tweepy.API(auth)
```

```python
twitname = "@marylony"
follower= api.followers_ids(twitname)
```

```python
usertweet = []
count = 0
```

```python
len(follower)
```

```python
for i in follower:
    error = False
    follower.remove(i)
    try:
        user = api.get_user(str(i))
    except:
        continue
    userjson = user._json
    id_user = userjson['id_str']
    status_count = userjson['statuses_count']
    if(status_count < 2000):
        continue
    print("count : " ,count)
    print("userid : ",id_user)
    cnt = 0
    page = 1
    while(True):
        try:
            if(cnt == 1000):
                break
            tweets = api.user_timeline(id_user, count = 200, page = page)
            print("page : ",page)
            for tweet in tweets:
                usertweet.append(json.dumps(tweet._json))
                cnt+=1
                if(cnt == 1000):
                    break
            page+=1
        except:
            error = True
            break
    if(error):
        continue
    count+=1
    print("toatl : ",len(usertweet))
    print("----------------------------------------------------")
    if count == 100:
        break;
```

```python
with open('tweets.json', 'w') as fp:
    json.dump(usertweet, fp)
```
